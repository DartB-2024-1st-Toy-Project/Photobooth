```python
# clean_column_value 함수 정의
def clean_column_value(value):
    value = re.sub(r'\([^)]*\)', '', value)  # 괄호와 괄호 안에 있는 글자 삭제
    value = re.sub(r'_측', '', value)  # '_측' 패턴 삭제
    value = value.replace('앞', '')  # '앞'이 포함된 경우 삭제
    if value.endswith('옆'):  # 마지막 글자가 '옆'일 때 '옆' 제거
        value = value[:-1]
    elif value.endswith('가'):  # 마지막 글자가 '가'일 때 '가' 제거
        value = value[:-1]
    elif value.endswith('측') and len(value) > 1:
        value = value[:-2]  # 마지막이 '측'으로 끝나면 두 번째 요소까지 선택
    
    parts = value.split()  # 띄어쓰기를 기준으로 문자열을 나눔
    if len(parts) >= 2:  # 두 번째 요소가 있으면 첫 번째와 두 번째 요소 선택
        return ' '.join(parts[:2])  # 두 개의 요소를 공백으로 연결하여 반환
    else:  # 두 번째 요소가 없으면 첫 번째 요소만 반환
        return parts[0]  # 첫 번째 요소 반환
```
