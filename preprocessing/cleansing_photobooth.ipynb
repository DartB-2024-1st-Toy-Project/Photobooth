```python
#가게 이름 갖고 할 건 아니니까
store_df.drop(columns=['지점명', '층정보', '호정보', '건물부번지', '건물관리번호', '건물명', '대지구분코드', '대지구분명', '시도코드', '시도명'], inplace=True)
```

```python
store_df
```

```python
# '주소' 컬럼을 공백을 기준으로 분리
store_df[['시', '구', '동', '그외주소']] = store_df['지번주소'].str.split(' ', 3).tolist()
```

```python
store_df['구_동'] = store_df['구'] + ' ' + store_df['동']
```

```python
#사진
#'상권업종중분류명'이 사진 촬영인 것. 상권업종소분류명이 사진촬영업인 거거
photo_shooting_df = store_df[store_df['상권업종중분류명'] == '사진 촬영']
set(photo_shooting_df['상호명'])
```

```python
photo_shooting_df.nunique()
```

```python
# '상호명'을 기준으로 그룹화하고 각 그룹별로 카운트하여 내림차순으로 정렬
count_by_store_name = store_df[store_df['상권업종중분류명'] == '사진 촬영'].groupby('상호명').size().reset_index(name='카운트').sort_values(by='카운트', ascending=False)
```

```python
brand_photoshop_name = count_by_store_name['상호명']
```

```python
# 'filtered_df'의 '상호명'이 'brand_photoshop'의 '상호명'으로 시작하는 경우 대체
for brand_name in brand_photoshop_name:
    store_df[store_df['상권업종중분류명'] == '사진 촬영'][store_df[store_df['상권업종중분류명'] == '사진 촬영']['상호명'].str.endswith('점')].loc[store_df[store_df['상권업종중분류명'] == '사진 촬영'][store_df[store_df['상권업종중분류명'] == '사진 촬영']['상호명'].str.endswith('점')]['상호명'].str.startswith(brand_name), '상호명'] = brand_name

# 결과 출력
print(store_df[store_df['상권업종중분류명'] == '사진 촬영'][store_df[store_df['상권업종중분류명'] == '사진 촬영']['상호명'].str.endswith('점')])
```
