```python
import pandas as pd
import requests

# 데이터프레임 순회하며 장소 검색 및 정보 저장
def search_places_and_add_info(df):
    # API 정보 설정
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    REST_API_KEY = "c93d281b676e49041098ce93d062477d"
    headers = {"Authorization": "KakaoAK {}".format(REST_API_KEY)}

    # 장소 정보를 추가할 리스트 생성
    place_info_list = []

    # 데이터프레임 순회하며 장소 검색 및 정보 추가
    for index, row in df.iterrows():
        keyword = row['장소']
        
        # 요청 파라미터 설정
        params = {
            "query": keyword
        }
        
        # API 호출
        response = requests.get(url, params=params, headers=headers)
        data = response.json()
        
        # 장소 정보 확인
        if 'documents' in data and data['documents']:  # 검색 결과가 비어있지 않은 경우
            for place in data['documents']:
                address = place.get('address_name')
                if address.startswith('서울'):  # 주소가 '서울'로 시작하는 경우
                    place_info = {
                        '이름': row['장소'],
                        '장소명': place.get('place_name'),
                        '주소': address,
                        '전화번호': place.get('phone')
                    }
                    place_info_list.append(place_info)
                    break  # 첫 번째로 찾은 장소 정보만 사용하고 반복문 종료
        else:
            print(f"{keyword}에 대한 검색 결과가 없습니다.")

    # 장소 정보를 데이터프레임으로 변환
    places_df = pd.DataFrame(place_info_list)
    
    return places_df


# 장소 검색 및 정보 추가
places_info_df = search_places_and_add_info(hotplace_df)

# 결과 출력
print(places_info_df)
```

```python
# places_info_df와 rent_df를 merge
merged_df = pd.merge(places_info_df, rent_df, how='left', left_on='이름', right_on='Unnamed: 2')
```

```python
import pandas as pd
import requests

def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    # 'KaKaoAK '는 그대로 두시고 개인키만 입력하세요.
    headers = {"Authorization": "KakaoAK c93d281b676e49041098ce93d062477d"}
    response = requests.get(url, headers=headers)
    api_json = response.json()
    if 'documents' in api_json and api_json['documents']:
        address_info = api_json['documents'][0]['address']
        lat = address_info['y']
        lng = address_info['x']
        return lat, lng
    else:
        return None, None

# '주소' 칼럼의 각 요소에 대해 위도와 경도 가져오기
lat_list = []
lng_list = []
for address in merged_df['주소']:
    lat, lng = get_location(address)
    lat_list.append(lat)
    lng_list.append(lng)

# 위도와 경도 정보를 데이터프레임에 추가
merged_df['위도'] = lat_list
merged_df['경도'] = lng_list

# 결과 확인
print(merged_df)
```
